```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Ganti URL dengan URL file CSV dari GitHub
url = 'https://raw.githubusercontent.com/taudataanalytics/eLearning/master/data/tips.csv'

# Baca file CSV dan simpan ke DataFrame
data = pd.read_csv(url)
data
```

```python
data.info()
```

## Data understanding

2. tip         : memberikan tambahan uang

4. smoker      : perokok atau tidak
5. day         : Hari apa yang dikunjungi

```python
# Mendapatkan kuartil pertama dan ketiga
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)

# Menghitung IQR
IQR = Q3 - Q1

# Menentukan batas atas dan batas bawah untuk outlier
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Mendeteksi outlier
outliers = ((data < lower_bound) | (data > upper_bound)).any(axis=1)

# Menampilkan data yang dianggap sebagai outlier
outlier_data = data[outliers]
print(outlier_data)
```
