```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# Ganti URL dengan URL file CSV dari GitHub
url = 'https://raw.githubusercontent.com/taudataanalytics/eLearning/master/data/tips.csv'

# Baca file CSV dan simpan ke DataFrame
data = pd.read_csv(url)
data
```

```python
# Mendapatkan kuartil pertama dan ketiga
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)

# Menghitung IQR
IQR = Q3 - Q1

# Menentukan batas atas dan batas bawah untuk outlier
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Mendeteksi outlier
outliers = ((data < lower_bound) | (data > upper_bound)).any(axis=1)

# Menampilkan data yang dianggap sebagai outlier
outlier_data = data[outliers]
outlier_data
```

Karena data yang outlier bukan data yang noise dan masih bisa dianalisis lebih lanjut maka dapat kita pisahkan terlebih dahulu

```python
# Misalkan 'data' adalah DataFrame yang sudah berisi data yang ingin kamu tinjau
# data = ...

# Mendapatkan kuartil pertama dan ketiga
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)

# Menghitung IQR
IQR = Q3 - Q1

# Menentukan batas atas dan batas bawah untuk outlier
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Membuat kondisi Boolean untuk menyaring data yang bukan outlier
not_outliers = ~((data < lower_bound) | (data > upper_bound)).any(axis=1)

# Mengambil data yang bukan outlier
data_no_outliers = data[not_outliers]
data_no_outliers
```

kemudian kita lihat sebaran datanya seperti apa setelah dipisahkan dengan outlier

```python
plt.hist(data_no_outliers['total_bill'])  # Ubah 'nama_kolom' dengan kolom yang ingin divisualisasikan

plt.xlabel('Nilai')
plt.ylabel('total bil')
plt.title('Histogram Data total bil')

plt.show()
```

```python
correlation_matrix = data_no_outliers.corr()

# Visualisasi matriks korelasi dengan heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriks Korelasi')

plt.show()
```
